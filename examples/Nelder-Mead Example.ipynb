```julia
using SpectraFit, Plots
```

```julia
experimental = get_experimental("B2O3FastCool.txt", 32.239);
```

```julia
result = fit_quadrupolar(experimental, 1, 100);
```

```julia
theoretical = generate_theoretical_spectrum(experimental, Quadrupolar(result.minimizer))
plot(experimental[:, 1], experimental[:, 2], label = "experimental")
theoretical_old = generate_theoretical_spectrum(experimental, Quadrupolar([5.4, 0.25, 0.12, 0.06, 1.0]))
plot!(experimental[:, 1], theoretical_old, width = 2, label = "old theoretical")
plot!(experimental[:, 1], theoretical, width = 2, label = "new theoretical", 
    title = "Theoretical vs. Experimental", xlabel = "Frequency (MHz)", ylabel = "Intensity")
```
